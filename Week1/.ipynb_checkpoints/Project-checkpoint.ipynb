

#Simple interest
r= float(input('what is the rate'))
t= float(input('what is the time'))
p= float(input('what is the principle'))
a= ((p * r * t)/100)
print('Your simple interest is:  ', a)



#Compound interest
r= float(input('what is the rate'))
t= float(input('what is the time'))
p= float(input('what is the principle'))
n= float(input('how many years is it compounded over:'))
a= p * (1 + (r/100)/n) ** (n*t)
print('The compound interest is: ', a)
