```python
import tkinter as tk
from tkinter import messagebox as msgbox
import pandas as pd

df = pd.read_csv("GIG-logistics.csv")

df["c_name"] = df["Name"].str.strip().str.lower()
df["c_dept"] = df["Department"].str.strip().str.lower()

def button_click():
    msgbox.showinfo("Info", "Welcome to GIG-Logistics employee Database")
    
    if msgbox.askyesno("Confirmation", "Do you want to continue?"):
        open_login()

def open_login():
    home.destroy() 

    global root, name_entry, dept_entry

    root = tk.Tk()
    root.title("Login Gateway")
    root.geometry("400x200")

    tk.Label(root, text="Name").pack()
    name_entry = tk.Entry(root)
    name_entry.pack()

    tk.Label(root, text="Department").pack()
    dept_entry = tk.Entry(root)
    dept_entry.pack()

    tk.Button(root, text="Submit", command=submit).pack(pady=10)

    root.mainloop()

def submit():
    name = name_entry.get().strip().lower()
    dept = dept_entry.get().strip().lower()

    if ((df["c_name"] == name) & (df["c_dept"] == dept)).any():

        team = df[df["c_dept"] == dept]["Name"].tolist()

    
        team = [t for t in team if t.lower() != name]

        msg = f"Welcome {name.title()}!\n\nOther members:\n"
        msg += "\n".join(team) if team else "No other members."

        msgbox.showinfo("Success", msg)

    else:
        msgbox.showerror("Error", "Invalid Name or Department")


home = tk.Tk()
home.title("Home Page")
home.geometry("400x150")

tk.Label(home, text="GIG Database\n").pack()

tk.Button(home, text="Enter!", command=button_click, fg="white", bg="black").pack()

home.mainloop()
```
