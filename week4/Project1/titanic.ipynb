```python
import pandas as pd
df= pd.read_csv("titanic.csv")
df
```

```python
df.info()
df.isnull().sum()
```

```python
#clean data
df.dropna(subset=['Embarked'], inplace=True)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df.drop(columns=["Cabin"] , inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#visualize the survivale rate showing more people died that survived 
#also shows that the average survival rate of women was more that that of men
sns.countplot(x="Survived", data=df)
plt.title("Survival Count (0 = died, 1 = survived)")
plt.show()

sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
#age distribution of passengers on the ship
#this show there where more middle aged people on the ship few young and old people
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```
