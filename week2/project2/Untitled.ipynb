```python
import numpy as np

def solve_equation():
    print("This code finds the root of polynomial")
    print("1. Quadratic ")
    print("2. Cubic ")
    print("3. Quartic")
    
    choice = input("Select Number")

    if choice == '1':
        print("Enter A,B,C")
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        coeffs = [a, b, c]
        
    elif choice == '2':
        print("Enter A,B,C,D")
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        d = float(input("Enter D: "))
        coeffs = [a, b, c, d]
        
    elif choice == '3':
        print("Enter A,B,C,D,E")
        a = float(input("Enter A: "))
        b = float(input("Enter B: "))
        c = float(input("Enter C: "))
        d = float(input("Enter D: "))
        e = float(input("Enter E: "))
        coeffs = [a, b, c, d, e]
          
    else:
        print("Invalid selection.")
        return

    roots = np.roots(coeffs)
    print("The roots are:")
    for i, root in enumerate(roots, 1):
        print(f"Root {i}: {np.round(root, 4)}")

    
    


solve_equation()

    
    
    
    


                  
```
